```python
import pandas as pd 
```

```python
movies_df = pd.read_csv('movies.csv',index_col=0)
movies_df.head()
```

```python
ratings_df = pd.read_csv('ratings.csv',index_col=0)
ratings_df.head()
```

```python
movies_rating_df = pd.merge(movies_df,ratings_df,on='movieId')
movies_rating_df.head()
```

```python
movies_by_average_rating = pd.DataFrame(movies_rating_df.groupby('title')['rating'].mean().sort_values(ascending=False))
movies_by_average_rating.head()
```

```python
movies_rating_df.groupby('title')['rating'].count()
```

```python
movies_by_average_rating['Count'] = pd.DataFrame(movies_rating_df.groupby('title')['rating'].count())
movies_by_average_rating.head()
```

```python
popular_movies = movies_by_average_rating[(movies_by_average_rating.rating > 3.0) & (movies_by_average_rating.Count > 100)]
```

```python
top_10 = popular_movies.head(10)
top_10
```
